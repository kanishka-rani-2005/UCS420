```python
import pandas as pd;
import numpy as np;
np.random.seed(102317165)
```

```python
sales_data = np.random.randint(1000, 5000, size=(12, 4))
# Create a DataFrame
columns = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df = pd.DataFrame(data=sales_data, columns=columns,index=months)
df.head()
```

```python
df.info
```

```python
df.describe()
```

```python
category_totals = df.sum(axis=0)
print("Total Sales per Category:\n", category_totals)

monthly_totals = df.sum(axis=1)
print("\nTotal Sales per Month:\n", monthly_totals)
```

```python
avg_growth_per_category = df.diff().mean()
print('Average Monthly Sales Growth per Category :\n',avg_growth_per_category)
```

```python
df['Total_Sales']=df.sum(axis=1)
df['Growth Rate'] = df['Total_Sales'].pct_change() * 100
df.head()
```

```python
rno=102317165

if rno%2==0:
    print("Even number")
    df['Electronics']=df['Electronics']*0.90
else:
    print("Odd number")
    df['Clothing']=df['Clothing']*0.85

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for column in ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']:
    plt.plot(df.index, df[column], marker='o', label=column)

plt.title('Monthly Sales Trends by Category')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python

df_melted = df[['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']].melt(var_name='Category', value_name='Sales')

plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Sales', data=df_melted, palette='Set2')
plt.title('Sales Distribution per Category')
plt.xlabel('Product Category')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]]) 


```python
array = np.array([[1, -2, 3],[-4, 5, -6]]) 


print(np.abs(array))

print('\nColumn wise Percentile.')
print(np.percentile(array,25,axis=0))
print(np.percentile(array,50,axis=0))
print(np.percentile(array,75,axis=0))

print('\nRow wise Percentile.')
print(np.percentile(array,25,axis=1))
print(np.percentile(array,50,axis=1))
print(np.percentile(array,75,axis=1))

```

```python

print('\nMean ')
print(np.mean(array,axis=1))
print(np.mean(array,axis=0))

print('\nMedian')
print(np.median(array,axis=1))
print(np.median(array,axis=0))

print('\nStandard Deviation')
print(np.std(array,axis=1))
print(np.std(array,axis=0))
```

Q.3 For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]). 


```python
a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0])

print("Floor values are :\n",np.floor(a))
print("Ceil values are :\n",np.ceil(a))
print("Truncated values are :\n",np.trunc(a))
print("Rounded values:\n",np.round(a,0))
```

Q.4 WAP to swap two elements in a list in Python Using a Temporary Variable. 


```python
my_list = [10, 20, 30, 40, 50]

i=int(input('enter first index '))
j=int(input('enter second index'))
print(my_list)

temp=my_list[i]
my_list[i]=my_list[j]
my_list[j]=temp

print(my_list)
```
