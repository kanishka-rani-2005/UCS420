Q.1 For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'),Find


i. Sum of all 

ii. Sum of all elements row-

iii. Sum of all elements column-wise

```python
import numpy as np;

gfg=np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print(gfg)

```

```python
#1.
sum=np.sum(gfg)
print(sum)
```

```python
#2.

for i in gfg :
    sum=np.sum(i)
    print(sum)

```

```python
# 2.
row_sum=np.sum(gfg,axis=1)
print(row_sum)
```

```python
#3.
col_sum=np.sum(gfg,axis=0)
print(col_sum)
```

Q.3 You are given a weekly sales dataset and need to perform various operations
using NumPy broadcasting.

a) Generate your unique sales dataset:

 Take the sum of the ASCII values of the initials of your first and last

name. Call this value X. (If your initials are A B → ASCII sum = 65
+ 66 = 131 → sales = [131, 181, 231, 281, 331].)

 Create a NumPy array sales with values [X, X+50, X+100, X+150,
X+200].

```python
first_name=input("enter your first name\n")
last_name=input("enter your last name\n")


X=ord(first_name[0])+ord(last_name[0])

print("Sum of ascii's is ",X)

```

```python
def createarray(X):
    array=np.array([X,X+50,X+100,X+150,X+200])
    return array


array=createarray(X)
```

b) Compute your personalized tax rate as ((X % 5) + 5) / 100.

 Use broadcasting to apply this tax rate to each sales value.

```python
def taxRate(X):
    return ((X%5)+5)/100


tax=taxRate(X)

taxed_sales = array * (1+tax)
print(taxed_sales)

```

c) Adjust sales based on discount:

 If sales < X+100, apply a 5% discount.

 If sales >= X+100, apply a 10% discount.

```python
def discounted_sales(array,X):
    discounted_array=np.where(array<X+100,array*0.95,array*0.90)
    return discounted_array

print(discounted_sales(array,X))
```

d) Expand sales data for multiple weeks:

 Create a 3×5 matrix representing three weeks of sales by stacking

sales three times using broadcasting.

 Increase sales by 2% per week using element-wise broadcasting.

```python
weekly_sales = np.vstack([array] * 3)  

print(weekly_sales)

weeks = np.arange(3).reshape(3, 1) 
weekly_sales = weekly_sales * (1.02 ** weeks)  

print("Weekly Sales Data:\n", weekly_sales)
```

Q4. Generate x values using np.linspace() from -10 to 10 with 100 points. Use

each function from the list below and compute y values using NumPy:

a) Y = x^2

b) Y = sin(x)

c) Y = e^x

d) Y = log(|x| + 1)

Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-10,10,100)
# print(x)

y1=x**2
y2=np.sin(x**2)
y3=np.exp(x)
y4=np.log(np.abs(x)+1)

plt.figure(figsize=(10,6))
```

```python

plt.plot(x,y1,color='blue',linestyle='-',label="function_1")
plt.title('Mathematical Function ', fontsize=20)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc='lower right')
plt.grid(True,linestyle='-',alpha=0.6)
```

```python
plt.plot(x,y2,color='green',linestyle='--',label="function_2")
plt.title('Mathematical Function ', fontsize=20)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc='upper left')
plt.grid(True,linestyle='--',alpha=0.7)
```

```python
plt.plot(x,y3,color='red',label="function_3",linestyle='--')
plt.title('Mathematical Function ', fontsize=20)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True,linestyle='-',alpha=0.8)
```

```python
plt.plot(x,y4,label='function_4',color='black')
plt.title('Mathematical Function ', fontsize=20)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.9)
```
