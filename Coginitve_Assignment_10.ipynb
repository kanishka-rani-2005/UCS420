3. Print and interpret the top 3 keywords from each text using TF-IDF.


```python
import numpy as np
def top_keywords(tfidf_vector, feature_names, top_n=3):
    sorted_indices = np.argsort(tfidf_vector.toarray()[0])[::-1]
    top_features = [(feature_names[i], tfidf_vector.toarray()[0][i]) for i in sorted_indices[:top_n]]
    return top_features
for idx, text in enumerate(texts):
    print(f"\nText {idx+1}: {text}")
    keywords = top_keywords(tfidf_matrix[idx], features_name)
    for word, score in keywords:
        print(f"{word}: {score:.4f}")

```

Q5. Write a short review for a product or service.
1. Use TextBlob or VADER to find polarity & subjectivity for each review.
2. Classify reviews into PosiƟve / Negative / Neutral.
3. Create a word cloud using the wordcloud library for all positive reviews.


```python
from textblob import TextBlob
from wordcloud import WordCloud

reviews = [
    "This phone has an amazing display and battery life.",
    "The customer service was terrible and unhelpful.",
    "Delivery was okay, but the packaging was damaged.",
    "Absolutely love the sound quality of these headphones!",
    "Worst experience ever. I would not recommend it."
]

postive_text=''
negative_text=''
for i ,review in enumerate(reviews):
  blob=TextBlob(review)
  polarity=blob.sentiment.polarity
  subjectivity=blob.sentiment.subjectivity

  if(polarity>0.1) :
    print(f"Review {i+1}: Positive having polarity : ",polarity)
    postive_text+=(review)
  elif(polarity<-0.1):
    print(f"Review {i+1}: Negative having polarity : ",polarity)
    negative_text+=(review)
  else:
    print(f"Review {i+1}: Neutral having polarity : ",polarity)

  print(f"Review {i+1}: Subjectivity :",subjectivity)
```

```python
import matplotlib.pyplot as plt
wordcloud = WordCloud().generate(postive_text)
plt.figure(figsize=(20, 4))
plt.imshow(wordcloud)
plt.axis('off')
plt.title("Word Cloud For Positive Reviews")
plt.show()
```

```python
wordcloud2=WordCloud().generate(negative_text)
plt.figure(figsize=(20,4))
plt.imshow(wordcloud2)
plt.axis('off')
plt.title('Word Cloud For Negative Reviews')
plt.show()
```

Q6. Choose your own paragraph (~100 words) as training data.
1. Tokenize text using Tokenizer() from keras.preprocessing.text
2. Create input sequences and build a simple LSTM or Dense model
3. Train the model and generate 2–3 new lines of text starƟng from any seed word you
provide.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical, pad_sequences
import numpy as np

text = """Machine learning is transforming how we interact with technology. From voice assistants to recommendation systems, algorithms are learning to understand and predict human behavior."""

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1
tokens = tokenizer.texts_to_sequences([text])[0]

input_sequences = []
for i in range(3, len(tokens)):
    input_sequences.append(tokens[i-3:i+1])

# Extracting features and labels
input_sequences = np.array(input_sequences)
X = input_sequences[:, :-1] # except last column
y = input_sequences[:, -1]  # only last column
y = to_categorical(y, num_classes=total_words)

# Building model
model = Sequential()
model.add(Embedding(total_words, 10, input_length=3))
model.add(LSTM(64))
model.add(Dense(total_words, activation='softmax'))

# compiling the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=300, verbose=0)
print("Model trained.")
```

```python
def generate_text(seed_text, next_words=3):
    for _ in range(next_words):
        # convert the text to sequence of token integers
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        # print(token_list)
        # if length is not 3 then it will left pad with zeros
        token_list = pad_sequences([token_list], maxlen=3)
        # print(token_list)
        # -1 for getting prediction from last layer
        # 0 means get the first and only
        predicted = np.argmax(model.predict(token_list, verbose=0), axis=-1)[0]

        for word, index in tokenizer.word_index.items():
            if index == predicted:
                seed_text += ' ' + word
                break
    return seed_text

print(generate_text("machine learning"))
print(generate_text("the future"))
```
