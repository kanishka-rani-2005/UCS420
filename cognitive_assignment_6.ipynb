 Q1. Ask the user to enter a value (e.g., M) for a mathematical function. ->Generate x values from -10 to 10 using np.linspace().

 ->Compute y values for:

 -> y= M⋅x^2

 ->y=M⋅sin(x)

 ->Plot both functions in a single figure:

 ->Use different colors and line styles.
 
 ->Add legend, grid, and title. 


```python
import numpy as np;
import matplotlib;
import matplotlib.pyplot as plt;
import seaborn as sns;
```

```python

M = float(input("Enter a value for M: "))
x=np.linspace(-10,10,20)
# print(x)

y1=M* x**2
y2=M*(np.sin(x))

plt.plot(x,y1,label="x v/s y1",color="red")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('X V/s Y1')
plt.show()
```

```python


plt.plot(x,y2,label="x v/s y2",color="blue")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('X V/s Y2')
plt.show()
```

```python
plt.plot(x,y1,label=r"$y=Mx^2$",color="red")
plt.plot(x,y2,label=r"$y=Msin(x)$",color="blue")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title(f'Y={M}x^2 y={M}sin(x)')
plt.legend(loc="upper right")
plt.grid(True,linestyle="--")
plt.show()
```

Q.2 Create a dataset of five subjects and scores.
->Convert it into a Pandas DataFrame.

-> Plot the scores using a Seaborn bar plot with:

-> Different colors for each bar.

-> Annotations on top of each bar.

-> Title, axis labels, and grid. 


```python
import pandas as pd

# Create dataset
data = {
    "Subject": ["Math", "Physics", "Chemistry", "Biology", "Computer Science"],
    "Score": [85, 90, 78, 88, 95]
}
```

```python
df=pd.DataFrame(data)
print(df)
```

<!-- ax.text(...)
This function adds text at a specific position on the plot.

First argument: p.get_x() + p.get_width() / 2 (X-position, centered on the bar).
Second argument: p.get_height() + 1 (Y-position, slightly above the bar).
Third argument: f'{p.get_height()}' (Text label, showing the value).
ha='center': Horizontally centers the text on the bar.
fontsize=12: Sets font size to 12.
fontweight='bold': Makes the text bold for better visibility. -->

```python
ax=sns.barplot(x='Subject',y='Score',data=df,palette="husl")

sns.set_style("whitegrid")


for i ,mark in enumerate(df['Score']):
    ax.text(i,mark+0.5,f"{mark}",ha='center',va='bottom',fontsize=10)

plt.title('Barplot')
plt.xlabel('SUBJECTS')
plt.ylabel('SCORES')
plt.grid(True,linestyle='--')
plt.legend(loc='lower right')
plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().

-> Create a 2x2 subplot layout displaying:

-> Line plot showing cumulative sum.

-> Scatter plot with random noise.

-> Customize with titles, axis labels, and grids. 


```python
roll_number=int(input('Enter your roll number as seed : '))

np.random.seed(roll_number)
data=np.random.randn(50)
print(data)

fig, ax = plt.subplots(2, 2, figsize=(15, 8))

ax[0][0].plot(np.cumsum(data),'b-o',label='Cumulative sum',color='red')
ax[0][0].set_title('Cumulative Sum Line Plot ')
ax[0][0].set_ylabel('Sum')
ax[0][0].set_xlabel('Index')
ax[0][0].legend(loc='upper left')
ax[0][0].grid(True,linestyle='--')

x = np.arange(50)
y = data + np.random.normal(0, 0.5, 50)

ax[0][1].scatter(x,y,marker='p',label='Noisy data')
ax[0][1].set_title('Scattered Plot with Noisy data ')
ax[0][1].set_ylabel('Value')
ax[0][1].set_xlabel('Index')
ax[0][1].legend(loc='upper left')
ax[0][1].grid(True,linestyle='--')



ax[1][0].axis('off')

ax[1][1].axis('off')

```

Q.4
Apply ‘seaborn’ library to do the following.

1. Read Total profit of all months and show it using a line plot.

2. Read all product sales data and show it using a multiline plot.

3. Plot bar chart for all the features/attributes.

```python
url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
df = pd.read_csv(url)

data=pd.DataFrame(df)
# print(data)

sns.set_style('whitegrid')

sns.lineplot(x=data['month_number'],y=data["total_profit"],label='Total profit',color='red',marker='o')
plt.title("Total Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.legend(loc='upper left')
plt.grid(True,linestyle='--')
plt.show()
```

```python

# Excluding 'Month' and 'Total Profit'

plt.figure(figsize=(20,8))
for i in data.columns[1:-1]:
    # print(i)
    sns.lineplot(x=data['month_number'],y=data[i],label=i,marker='o',linewidth=4)


plt.title("Product sales over month")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
